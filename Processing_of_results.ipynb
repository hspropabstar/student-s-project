```python
#Автоматизация обработки результатов, данные хранились в множестве папок. Нужно было получить значения параметров,
#сделать отчёт в ворд. Также сделать сводную таблицу в экселе. Интересная задача для сокращения затраченного времени на проект.
import os
import re
import pandas as pd
excel = pd.read_excel(r'Был путь к экселю')
column = 1
for i in os.scandir(r'Был путь в корневую папку, где хранились данные'):
    split = re.findall('[0-9]*[.][0-9]+',i.name)
    for j in range(0,15):
        excel[column][j] = split[j]
    column += 1
excel.to_excel(r'Заносим значение параметров в сводную таблицу')
```

```python
#Создание скрипта для получения картинок в программе Tecplot360. Автоматизация получения этих самых картинок на 26 папок, 
#В общей сумме вышел отчёт на 200 страниц Word.
import os
import re
#script - наш макрос
from script import script
names = []
Path_picture = []
#Создание папок в директории, добавление в names нужных plt(статик и последняя итерация). 
for i,j,k in os.walk(r'Также путь в корневую папку'):
    k.sort()
    if j: 
        for path in j:
            name_path = f'Создание упорядоченных папок, чтобы потом собрать там все картинки{path}'
            Path_picture.append(name_path)
            if not os.path.exists(name_path):
                os.makedirs(name_path)
    try:
        names.append([i + '\\' + k[-2], i + '\\' + k[-6]])
    except:
            pass
i=0       
#Создание макроса с путями, которые нужны и выполнение его
for path,picture in zip(names,Path_picture):
    i+=1 
    print(i)
    print('Current: ' + picture)
    new_names = f'"{path[0]}" "{path[1]}"'
    new_s = script.replace('Path_to_two_files', new_names)
    new_s = new_s.replace('Path_picture', picture)
    output_file = open(f'{picture}/macros.mcr', 'w')
    output_file.write(new_s)
    output_file.close()
    os.system('TEC360 '+ picture + r'\macros.mcr')
        
        
        


```

```python
import os
import re
import pandas as pd
row = 18
row1 = 22
min_strain_ijk = []
gap_i = []
for i,j,k in os.walk(r'Путь в корневую папку'):
    if 'myfile.csv' in k:
        ds = pd.read_csv(f"{i}\\GAP_I.csv",sep = ',')
        ds.columns = ['0','1']
        df = pd.read_csv(f"{i}\\myfile.csv")
        excel = pd.read_excel(r'Сводная таблица')
        min_strain_ijk.append((round(min(df.min()),4)))
        gap_i.append(ds[ds['0']>1.2])
for j in range(26):
    excel[j+1][row] = min_strain_ijk[j]
    excel[j+1][row1] = gap_i[j]
```
